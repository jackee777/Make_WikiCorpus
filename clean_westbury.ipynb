```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import nltk
words = set([word.lower() for word in nltk.corpus.words.words()])
total_words = 0
window = 5
progress_per = 10000
with open("WestburyLab/WestburyLab.Wikipedia.Corpus.txt", "r", encoding="utf-8") as f:
    with open("WestburyLab/WestburyLab.txt", "w") as fw:
        for sentence_no, sentence in enumerate(f):
            sentence = word_tokenize(sentence)

            new_review = []
            for token in sentence:
                new_token = regex.sub(u'', token)
                if not new_token == u'':
                    new_review.append(new_token)
                    
            if len(new_review) <= window or \
                np.sum([1 if token in words else 0 for token in new_review])/len(new_review) <= 0.5:
                continue

            if sentence_no % progress_per == 0:
                #print(sentence_no, " ".join(new_review))
                print(
                    "PROGRESS: at sentence #%i, processed %i words",
                    sentence_no, total_words
                )
            fw.write(" ".join(new_review)+"\n")
```

```python
sentence_no = -1
total_words = 0
min_reduce = 1
progress_per = 10000
vocab = collections.defaultdict(int)
char_vocab = collections.defaultdict(int)
checked_string_types = 0
with open("WestburyLab/WestburyLab.Wikipedia.Corpus.txt", "r", encoding="utf-8") as f:
    for sentence_no, sentence in enumerate(f):
        sentence = sentence.split()
        if not checked_string_types:
            if isinstance(sentence, str):
                print(
                    "Each 'sentences' item should be a list of words (usually unicode strings). "
                    "First item here is instead plain %s.",
                    type(sentence)
                )
            checked_string_types += 1
        if sentence_no % progress_per == 0:
            print(
                "PROGRESS: at sentence #%i, processed %i words, keeping %i word types",
                sentence_no, total_words, len(vocab)
            )

        for word in sentence:
            for j in word:
                char_vocab[j] += 1
            vocab[word] += 1
        total_words += len(sentence)
```
