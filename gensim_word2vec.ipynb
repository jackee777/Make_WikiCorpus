```python
import gensim
import logging
import string

#logging.basicConfig(format='%(asctime)s: % (levelname) s: % (message)s', level=logging.INFO)

"""
https://machinelearningmastery.com/clean-text-machine-learning-python/
"""


def read_input():
    table = str.maketrans("", "", string.punctuation)
    documents = []
    with open("WestburyLab/WestburyLab_English.txt", "r") as f:
        for i, line in enumerate(f):
            if ((i+1) % 10000 == 0):
                #print("read {0} reviews".format(i))
                logging.info("read %d reviews" % i)
                #yield
            document = line.split()
            if len(document) <= 1:
                continue
            documents.append([word.translate(table) for word in document])
        
        return documents
            #return line.translate(table).lower()

documents = read_input()
print(documents)
```
