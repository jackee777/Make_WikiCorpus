```python
import numpy as np
import random
random.seed(777)
```

```python
with open("WestburyLab/WestburyLab_English_50percent.txt", "r") as f:
    text = f.readlines()
```

```python
text[:10]
```

```python
def remove_word(document, target):
    return [word for word in document if word != target]
```

```python
def search_context(document, target):
    max_len = len(document)
    window = 5
    candidates = [i for i, word in enumerate(document) if word == target]
    bag_of_words = []
    for i in candidates:
        start = i - window
        end = i + window + 1
        if start < window:
            start = 0
        if end > max_len:
            end = max_len
        bag_of_words.append(remove_word(document[start:i]+document[i+1:end], target))
        #bag_of_words.append(" ".join(document[start:i]+document[i+1:end]))
        #bag_of_words.append(" ".join(document[start:end]).replace(target, ""))
    return bag_of_words
```

```python
target = "mouse"
bag_of_words = []
for document in text:
    document = document.replace("\n", "").split()
    bow = search_context(document, target)
    
    if bow != []:
        bag_of_words.extend(bow)
bag_of_words
```

```python
word2idx = dict()
for word in set([word for words in bag_of_words for word in words]):
    word2idx[word] = len(word2idx)
word2idx
```

```python
def bag_of_array(bag_of_words):
    bag_of_array = np.zeros((len(bag_of_words), len(word2idx)))
    for i,words in enumerate(bag_of_words):
        index = [word2idx[word] for word in words]
        bag_of_array[i][index] = 1
    return bag_of_array

boa = bag_of_array(bag_of_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
boa.shape
```

```python
k_num = 5
model = KMeans(n_clusters=k_num, init='k-means++', max_iter=100, n_init=1)
pred = model.fit_predict(boa)
pred
```

```python
cluster_index = []
for i in range(k_num):
    cluster_index.append([j for j,cluster in enumerate(pred) if cluster == i])
cluster_index
```

```python
print("number of cluster", np.max(pred))
print("number of sentences", len(pred))
```
